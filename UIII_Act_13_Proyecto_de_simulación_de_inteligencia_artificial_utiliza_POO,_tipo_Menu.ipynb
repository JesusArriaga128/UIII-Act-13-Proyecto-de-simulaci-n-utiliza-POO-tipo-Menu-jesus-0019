( Clase prestamo personal )

```python
class Prestamo_Personal:
    # Constructor de la clase donde se definen los 5 atributos
    def __init__(self, id_prestamo, monto_solicitado, tasa_interes_anual, plazo_meses):
        # 1. Identificador único del préstamo
        self.Id_Prestamo = id_prestamo
        # 2. Cantidad de dinero prestada inicialmente
        self.Monto_Solicitado = monto_solicitado
        # 3. Porcentaje de interés cobrado al año (ej: 0.15 para 15%)
        self.Tasa_Interes_Anual = tasa_interes_anual
        # 4. Duración del préstamo en meses
        self.Plazo_Meses = plazo_meses
        # 5. Dinero que falta por pagar (inicia siendo el total solicitado)
        self.Saldo_Pendiente = monto_solicitado

    # --- LAS 5 FUNCIONES (MÉTODOS) DENTRO DE LA CLASE ---

    # Función 1: Calcula el pago fijo mensual usando la fórmula de amortización
    def calcular_Cuota_Mensual(self):
        tasa_mensual = self.Tasa_Interes_Anual / 12
        if tasa_mensual > 0:
            cuota = (self.Monto_Solicitado * tasa_mensual) / (1 - (1 + tasa_mensual) ** -self.Plazo_Meses)
        else:
            cuota = self.Monto_Solicitado / self.Plazo_Meses
        return round(cuota, 2)

    # Función 2: Registra un pago hecho por el cliente y lo resta de la deuda
    def registrar_Abono(self, monto_pago):
        if monto_pago <= self.Saldo_Pendiente:
            self.Saldo_Pendiente -= monto_pago
            print(f"\n[ÉXITO] ¡Abono de ${monto_pago} registrado con éxito!")
        else:
            print(f"\n[AVISO] El pago excede la deuda. Se liquidó el saldo con ${round(self.Saldo_Pendiente, 2)}")
            self.Saldo_Pendiente = 0

    # Función 3: Aplica el cobro de interés mensual sobre el saldo actual
    def aplicar_Interes_Mes(self):
        tasa_mensual = self.Tasa_Interes_Anual / 12
        interes_generado = self.Saldo_Pendiente * tasa_mensual
        self.Saldo_Pendiente += interes_generado
        print(f"\n[SISTEMA] Interés mensual aplicado: +${round(interes_generado, 2)}")

    # Función 4: Muestra en pantalla el estatus completo del préstamo
    def consultar_Estado(self):
        print("\n=====================================")
        print("      ESTADO DEL PRÉSTAMO PERSONAL    ")
        print("=====================================")
        print(f"ID del Préstamo: {self.Id_Prestamo}")
        print(f"Monto Original: ${self.Monto_Solicitado:,.2f}")
        print(f"Plazo Total: {self.Plazo_Meses} meses")
        print(f"Tasa de Interés: {self.Tasa_Interes_Anual * 100}% anual")
        print(f"Saldo Pendiente Actual: ${round(self.Saldo_Pendiente, 2):,.2f}")
        print("=====================================\n")

    # Función 5: Calcula cuánto costaría pagar toda la deuda hoy (sin intereses futuros)
    def liquidar_Anticipado(self):
        print(f"\n[LIQUIDACIÓN] Para cancelar el préstamo '{self.Id_Prestamo}' hoy mismo, el pago neto es de: ${round(self.Saldo_Pendiente, 2):,.2f}")
        return self.Saldo_Pendiente


# --- BLOQUE PRINCIPAL CON VALIDACIÓN ANTIFALLOS (TRY-EXCEPT) ---
print("--- SISTEMA DE REGISTRO DE PRÉSTAMOS (CBTIS 128) ---")
print("Por favor, introduce los datos del nuevo préstamo:\n")

# 1. Captura de ID
id_in = input("1. Introduce el ID o Código del préstamo (ej: PREST-01): ").strip()

# 2. Captura y Validación de Monto
while True:
    try:
        monto_raw = input("2. ¿Cuál es el Monto Solicitado? $: ")
        monto_in = float(monto_raw.replace("$", "").replace(",", "").strip())
        if monto_in <= 0:
            print("[ERROR] El monto debe ser un número mayor a cero.")
            continue
        break # Si todo sale bien, rompe el ciclo y continúa
    except ValueError:
        print("[ERROR] Entrada inválida. Por favor ingresa solo números válidos (Ej: 10000).")

# 3. Captura y Validación de Tasa de Interés (Fila corregida con Try-Except)
while True:
    try:
        tasa_raw = input("3. Introduce la Tasa de Interés Anual en porcentaje (ej. 12 o 12%): ")
        tasa_anual_in = float(tasa_raw.replace("%", "").strip())
        if tasa_anual_in < 0:
            print("[ERROR] La tasa de interés no puede ser negativa.")
            continue
        tasa_decimal = tasa_anual_in / 100
        break # Rompe el ciclo si la conversión numérica fue exitosa
    except ValueError:
        print("[ERROR] Entrada inválida en la tasa. Por favor ingresa solo el número (Ej: 12 o 12.5).")

# 4. Captura y Validación de Plazo
while True:
    try:
        plazo_raw = input("4. ¿A cuántos meses se pagará el préstamo?: ")
        plazo_in = int(plazo_raw.strip())
        if plazo_in <= 0:
            print("[ERROR] El plazo debe ser de al menos 1 mes.")
            continue
        break
    except ValueError:
        print("[ERROR] Entrada inválida. Por favor ingresa un número entero de meses (Ej: 12).")


# --- EJECUCIÓN DEL PROGRAMA ---
prestamo_usuario = Prestamo_Personal(id_in, monto_in, tasa_decimal, plazo_in)
prestamo_usuario.consultar_Estado()

cuota_estimada = prestamo_usuario.calcular_Cuota_Mensual()
print(f"--> La cuota mensual calculada para este préstamo es de: ${cuota_estimada}")

print("\n--- SIMULACIÓN DE OPERACIÓN ---")

# Validación para el abono final
while True:
    try:
        pago_raw = input(f"¿Cuánto dinero desea abonar el cliente en este momento? (Sugerido: ${cuota_estimada}): ")
        pago_cliente = float(pago_raw.replace("$", "").replace(",", "").strip())
        if pago_cliente < 0:
            print("[ERROR] El abono no puede ser una cantidad negativa.")
            continue
        break
    except ValueError:
        print("[ERROR] Entrada inválida. Introduce una cantidad de dinero válida.")

prestamo_usuario.registrar_Abono(pago_cliente)
prestamo_usuario.consultar_Estado()
```
